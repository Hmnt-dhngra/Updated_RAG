### What is LLM? 

LLMs are essentially deep learning models which are basically collection of weights and biases. 

Formally, 
An LLM is a type of artificial intelligence trained on huge amounts of text so it can understand and generate human-like language.

Examples of LLMs include
- ChatGPT
- GPT - 4
- Gemini
- LLaMA


Most LLMs are built using a deep learning architecture called **Transformers**. 
Transformers usually follow Encoder - Decoder architecture. 

GPT follows Decoder architecture

In NLP, there are 2 stages of using Language models. 

Think of them like 
- Upstream : Training the brain of the model. 
- Downstream : Using the brain of the model to do real life tasks. 

Types of NLP tasks : 
- Upstream Task : Used to train the language model so that it learns the language patterns. They usually happen during pretraining of models like BERT, GPT, RoBERTa. Masking
- Downstream Task : Real world applications that use the pretrained models. Examples : Sentiment Analysis, Text Classification, Summarisation, Named Entity Recognition, POS, etc. 


RAG will be all about downstream tasks. 


Base model /Foundational model + Custom Data => Fine tuning Task
- SLM = any model having less than 1 B parameters
- LLM = more than 1 B parameters

Problems with SLM/LLMs
- Knowledge cutoff : The model usually have a training cutoff date beyond which the LLM/SLMs won't be able to generate correct information. 
- Lack of Information : Mainly domain specific. 
- Lack of private data : Won't have your organisational data. You can't train easily on your organisational data as it is resource intensive. 
- Traceability : Can't specifically cite the resources for specific responses. 
- Probabilistic output generation : The output can change basis the temperature requirements you give.
