In the CLI , you can get a specific model by running the model name using ollama like : 
- ollama pull embeddinggemma 
- ollama pull gemma3
- ollama list -> This gives you the models downloaded/available on your system. 
- ollama run gemma3 --> This command initiates the model to be run on CLI and there after you can ask you generic questions. 
- /bye --> This closes the chat window in your CLI. 

Ultimately you need atleast 1 embeddings model and 1 thinking model.
